```python
import matplotlib.pyplot as plt

#Function for parsing Log files
def get_loss_values(log):
    lines = log.split('\n')[5:-2]
    train_loss = [float(line.split()[11]) for line in lines]
    val_loss = [float(line.split()[-1]) for line in lines]

    return train_loss, val_loss

#Set the experiment path we want to load
exp_path = 'experiments/en-de(27.02.2024|19.07.58)'

#Read Log file
with open(f'{exp_path}/logging.log', 'r') as f:
    log = f.read()

#Get Loss history
train_loss, val_loss = get_loss_values(log)

plt.plot(train_loss, label = 'Train')
plt.plot(val_loss, label = 'Validation')
plt.legend()
plt.show()
```
